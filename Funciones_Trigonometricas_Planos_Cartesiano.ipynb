```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Crear un ángulo en radianes
angulo = np.pi/4  # 45 grados en radianes

# Calcular el seno, coseno y tangente del ángulo
seno = np.sin(angulo)
coseno = np.cos(angulo)
tangente = np.tan(angulo)

# Imprimir los resultados
print(f"Ángulo: {angulo} radianes")
print(f"Seno: {seno}")
print(f"Coseno: {coseno}")
print(f"Tangente: {tangente}")

# Dibujar el ángulo en el plano cartesiano
plt.figure(figsize=(5, 5))
plt.plot([0, np.cos(angulo)], [0, np.sin(angulo)], 'b', linewidth=2)
plt.plot([np.cos(angulo), np.cos(angulo)], [0, np.sin(angulo)], 'r--')
plt.plot([0, np.cos(angulo)], [0, 0], 'g--')
plt.text(np.cos(angulo)/2, -0.1, 'Coseno', color='b')
plt.text(np.cos(angulo)+0.1, np.sin(angulo)/2, 'Seno', color='b')
plt.text(np.cos(angulo)+0.1, -0.1, 'Tangente', color='r')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Ángulo en el Plano Cartesiano')
plt.grid(True)
plt.show()
```

```python
# Coordenadas de los puntos del triángulo
x = [0, 3, 0]
y = [0, 0, 4]

# Dibujar los lados del triángulo
plt.plot(x + [x[0]], y + [y[0]], marker='o')

# Etiquetar los vértices
plt.text(0, 0, 'O', ha='right')
plt.text(3, 0, 'A', ha='left')
plt.text(0, 4, 'B', va='bottom')

# Etiquetar los lados
plt.text(1.5, 0, '3', ha='center', va='top')
plt.text(0, 2, '4', ha='right', va='center')

# Configuración adicional del gráfico
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Triángulo Rectángulo')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Coordenadas del punto
x_punto = 2
y_punto = 3

# Coordenadas de la línea horizontal
x_linea = [0, 4]
y_linea = [1, 1]

# Dibujar el punto
plt.plot(x_punto, y_punto, 'ro', label='Punto')

# Dibujar la línea horizontal
plt.plot(x_linea, y_linea, 'b-', label='Línea Horizontal')

# Proyección ortogonal
plt.plot([x_punto, x_punto], [y_punto, 1], 'g--', label='Proyección Ortogonal')

# Etiquetas de puntos
plt.text(x_punto, y_punto + 0.2, 'Punto (2, 3)', ha='center')
plt.text(1, 1.2, 'Línea Horizontal', ha='center')
plt.text(x_punto + 0.2, 1, 'Proyección Ortogonal', va='center', rotation=90)

# Configuración adicional del gráfico
plt.xlim(-1, 5)
plt.ylim(-1, 4)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Proyección Ortogonal')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
# Coordenadas del vector
x_vector = 3
y_vector = 4

# Dibujar el vector
plt.quiver(0, 0, x_vector, y_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Vector')

# Proyección en el eje x
plt.quiver(0, 0, x_vector, 0, angles='xy', scale_units='xy', scale=1, color='r', label='Proyección en eje x')

# Etiquetas de los puntos
plt.text(x_vector, y_vector, 'Vector (3, 4)', ha='right', va='bottom')
plt.text(x_vector / 2, -0.5, 'Proyección en eje x', ha='right')

# Configuración adicional del gráfico
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Proyección de un Vector en el Eje X')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
# Crear un conjunto de ángulos
angulos = np.linspace(0, 2*np.pi, 100)  # Ángulos de 0 a 2*pi

# Calcular los valores de seno, coseno y tangente para cada ángulo
seno = np.sin(angulos)
coseno = np.cos(angulos)
tangente = np.tan(angulos)

# Crear gráficos para seno, coseno y tangente
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(angulos, seno, label='Seno')
plt.xlabel('Ángulo (x)')
plt.ylabel('Valor')
plt.title('Seno(x)')

plt.subplot(132)
plt.plot(angulos, coseno, label='Coseno', color='orange')
plt.xlabel('Ángulo (x)')
plt.ylabel('Valor')
plt.title('Coseno(x)')

plt.subplot(133)
plt.plot(angulos, tangente, label='Tangente', color='green')
plt.xlabel('Ángulo (x)')
plt.ylabel('Valor')
plt.title('Tangente(x)')

plt.tight_layout()
plt.show()
```
