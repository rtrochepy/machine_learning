```python
# Puntos de los vértices del triángulo rectángulo
A = (0, 0)
B = (3, 0)
C = (0, 4)

# Dibuja el triángulo
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], 'bo-')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('Triángulo Rectángulo')
plt.show()
```

```python
# Crear un conjunto de ángulos en grados
angulos_grados = np.linspace(0, 360, 361)  # Ángulos de 0 a 360 grados

# Convertir los ángulos a radianes (necesario para las funciones trigonométricas)
angulos_radianes = np.radians(angulos_grados)

# Calcular el seno, el coseno y la tangente de los ángulos en radianes
seno = np.sin(angulos_radianes)
coseno = np.cos(angulos_radianes)
tangente = np.tan(angulos_radianes)

# Graficar las funciones trigonométricas
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.plot(angulos_grados, seno)
plt.xlabel('Ángulo (grados)')
plt.ylabel('sin(θ)')
plt.title('Seno')

plt.subplot(132)
plt.plot(angulos_grados, coseno)
plt.xlabel('Ángulo (grados)')
plt.ylabel('cos(θ)')
plt.title('Coseno')

plt.subplot(133)
plt.plot(angulos_grados, tangente)
plt.xlabel('Ángulo (grados)')
plt.ylabel('tan(θ)')
plt.ylim(-10, 10)  # Limitar el rango de la tangente para que el gráfico sea legible
plt.title('Tangente')

plt.tight_layout()
plt.show()
```

```python
# Valores calculados de las funciones trigonométricas
angulo_grados = 36.87  # Este es el ángulo θ en grados, aproximadamente

# Crear una tabla con los valores
tabla = [['Función', 'Valor'],
         ['Seno (sin)', 3/5],
         ['Coseno (cos)', 4/5],
         ['Tangente (tan)', 3/4],
         ['Cosecante (csc)', 5/3],
         ['Secante (sec)', 5/4],
         ['Cotangente (cot)', 4/3]]

# Mostrar la tabla
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
ax.table(cellText=tabla, loc='center', cellLoc='center')

plt.title(f'Funciones Trigonométricas para θ ≈ {angulo_grados}°')
plt.show()
```

```python
# Coordenadas de los puntos M, N y O
M = (0, 0)  # Origen (0, 0)
N = (0, 4)  # Cateto opuesto (0, 4)
O = (3, 0)  # Cateto adyacente (3, 0)

# Dibujar los segmentos de los catetos
plt.plot([M[0], N[0]], [M[1], N[1]], label='Cateto opuesto (MN)', color='b')
plt.plot([M[0], O[0]], [M[1], O[1]], label='Cateto adyacente (MO)', color='g')

# Dibujar la hipotenusa
plt.plot([N[0], O[0]], [N[1], O[1]], label='Hipotenusa (NO)', color='r')

# Etiquetar los puntos
plt.text(M[0], M[1], 'M', fontsize=12, ha='right')
plt.text(N[0], N[1], 'N', fontsize=12, ha='left')
plt.text(O[0], O[1], 'O', fontsize=12, ha='right')

# Configurar ejes y etiquetas
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Configurar la leyenda
plt.legend()

# Mostrar el gráfico
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
# Longitudes de los catetos
cateto_opuesto = 4
cateto_adyacente = 3

# Usar el Teorema de Pitágoras para calcular la hipotenusa
hipotenusa = math.sqrt(cateto_opuesto**2 + cateto_adyacente**2)

# Imprimir el resultado
print(f"La hipotenusa es: {hipotenusa}")
```

```python
# Ángulo θ en radianes
theta = np.deg2rad(30)  # Convertir 30 grados a radianes

# Definir las longitudes de los catetos y la hipotenusa
cateto_opuesto = 1
cateto_adyacente = cateto_opuesto / np.tan(theta)  # Calcular el cateto adyacente
hipotenusa = np.sqrt(cateto_opuesto**2 + cateto_adyacente**2)

# Crear un triángulo rectángulo
triangle = plt.Polygon([[0, 0], [cateto_adyacente, 0], [0, cateto_opuesto]], closed=True, edgecolor='b', fill=False)

# Configurar el gráfico
fig, ax = plt.subplots()
ax.add_patch(triangle)
plt.plot([0, cateto_adyacente], [0, 0], 'r--')
plt.plot([0, 0], [0, cateto_opuesto], 'g--')
plt.text(cateto_adyacente/2, -0.1, 'Cateto Adyacente', fontsize=12, ha='center')
plt.text(-0.1, cateto_opuesto/2, 'Cateto Opuesto', fontsize=12, va='center', rotation='vertical')
plt.text(cateto_adyacente/2, cateto_opuesto/2, 'Hipotenusa', fontsize=12, ha='right')

# Configurar ejes y límites
plt.xlim(-0.5, cateto_adyacente + 0.5)
plt.ylim(-0.5, cateto_opuesto + 0.5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar el gráfico
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Triángulo Rectángulo para θ = 30 grados')

plt.show()
```

```python
# Ángulo a en radianes
a = np.deg2rad(30)  # Convertir 30 grados a radianes

# Valores de las funciones trigonométricas
sin_a = np.sin(a)
cos_a = np.cos(a)
tan_a = np.tan(a)
csc_a = 1 / sin_a
sec_a = 1 / cos_a
cot_a = 1 / tan_a

# Etiquetas para las funciones trigonométricas
labels = ['sin(a)', 'cos(a)', 'tan(a)', 'csc(a)', 'sec(a)', 'cot(a)']

# Valores correspondientes a las funciones trigonométricas
values = [sin_a, cos_a, tan_a, csc_a, sec_a, cot_a]

# Crear un gráfico de barras
plt.bar(labels, values, color=['blue', 'red', 'green', 'blue', 'red', 'green'])

# Configurar ejes y etiquetas
plt.ylabel('Valor de la función')
plt.title('Funciones trigonométricas y sus reciprocas para a = 30 grados')

# Mostrar el gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
