```python
import matplotlib.pyplot as plt
```

```python
# Coordenadas del punto
x = 3
y = 6

# Crear el gráfico del plano cartesiano
plt.figure(figsize=(6, 6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Graficar el punto
plt.plot(x, y, 'ro')  # 'ro' significa que se utilizará un punto rojo

# Etiquetar los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Establecer límites de los ejes
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Mostrar el gráfico
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Plano Cartesiano')
plt.show()
```

```python
# Coordenadas de los puntos
puntos = {
    "P1": (2, 3),
    "P2": (3, 6),
    "P3": (-2, -3),
    "P4": (-3, -6),
    "P5": (-2, 3),
    "P6": (3, -6),
    "P7": (2, -3),
    "P8": (-3, 6),
    "P9": (6, 6),
    "P10": (-6, -6),
}

# Crear el gráfico del plano cartesiano
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Graficar los puntos
for punto, coordenadas in puntos.items():
    x, y = coordenadas
    plt.plot(x, y, 'ro')  # 'ro' significa que se utilizará un punto rojo
    plt.text(x, y, punto, fontsize=12, ha='right')

# Etiquetar los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Establecer límites de los ejes
plt.xlim(-7, 7)
plt.ylim(-7, 7)

# Mostrar el gráfico
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Plano Cartesiano con Puntos')
plt.show()
```
