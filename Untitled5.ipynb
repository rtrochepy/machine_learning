```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Puntos de los vértices del triángulo rectángulo
A = (0, 0)
B = (3, 0)
C = (0, 4)

# Dibuja el triángulo
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], 'bo-')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('Triángulo Rectángulo')
plt.show()
```

```python
# Puntos de los vértices del triángulo obtusángulo
A = (1, 1)
B = (4, 1)
C = (2, 4)

# Dibuja el triángulo
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], 'bo-')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('Triángulo Obtusángulo')
plt.show()
```

```python
# Puntos de los vértices del triángulo acutángulo
A = (1, 2)
B = (4, 2)
C = (2, 4)

# Dibuja el triángulo
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], 'bo-')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('Triángulo Acutángulo')
plt.show()
```

```python
# Crear un conjunto de ángulos en grados
angulos_grados = np.linspace(0, 360, 361)  # Ángulos de 0 a 360 grados

# Convertir los ángulos a radianes (necesario para las funciones trigonométricas)
angulos_radianes = np.radians(angulos_grados)

# Calcular el seno, el coseno y la tangente de los ángulos en radianes
seno = np.sin(angulos_radianes)
coseno = np.cos(angulos_radianes)
tangente = np.tan(angulos_radianes)

# Graficar las funciones trigonométricas
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.plot(angulos_grados, seno)
plt.xlabel('Ángulo (grados)')
plt.ylabel('sin(θ)')
plt.title('Seno')

plt.subplot(132)
plt.plot(angulos_grados, coseno)
plt.xlabel('Ángulo (grados)')
plt.ylabel('cos(θ)')
plt.title('Coseno')

plt.subplot(133)
plt.plot(angulos_grados, tangente)
plt.xlabel('Ángulo (grados)')
plt.ylabel('tan(θ)')
plt.ylim(-10, 10)  # Limitar el rango de la tangente para que el gráfico sea legible
plt.title('Tangente')

plt.tight_layout()
plt.show()
```

```python
# Valores calculados de las funciones trigonométricas
angulo_grados = 36.87  # Este es el ángulo θ en grados, aproximadamente

# Crear una tabla con los valores
tabla = [['Función', 'Valor'],
         ['Seno (sin)', 3/5],
         ['Coseno (cos)', 4/5],
         ['Tangente (tan)', 3/4],
         ['Cosecante (csc)', 5/3],
         ['Secante (sec)', 5/4],
         ['Cotangente (cot)', 4/3]]

# Mostrar la tabla
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
ax.table(cellText=tabla, loc='center', cellLoc='center')

plt.title(f'Funciones Trigonométricas para θ ≈ {angulo_grados}°')
plt.show()
```
