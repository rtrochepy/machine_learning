```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Crear una grilla 5x5
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(6):
    ax.plot([i, i], [0, 5], 'k-')
    ax.plot([0, 5], [i, i], 'k-')
```

```python
# Dibujar las longitudes
# Horizontales
for i in range(4):
    ax.plot([0, i + 1], [0, 0], 'b-', lw=2)
    ax.text((i + 1)/2, -0.3, f"{i + 1}", color="blue")
```

```python
# Verticales (solo etiquetamos para no sobreponer las etiquetas)
for i in range(4):
    ax.text(-0.5, (i + 1)/2, f"{i + 1}", color="blue")
```

```python
# Diagonales
diagonales = [
    (1, 1),
    (2, 2),
    (3, 3),
    (4, 4),
    (1, 2),
    (1, 3),
    (1, 4),
    (2, 3),
    (2, 4),
    (3, 4)
]
```

```python
for dx, dy in diagonales:
    length = np.sqrt(dx**2 + dy**2)
    ax.plot([0, dx], [0, dy], 'r-', lw=2)
    ax.text(dx/2, dy/2, f"{length:.2f}", color="red")
```

```python
ax.set_aspect('equal')
ax.axis('off')
plt.show()
```

```python
def draw_geoboard():
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 5)

    # Dibujamos los puntos del geoboard
    for x in range(6):
        for y in range(6):
            ax.plot(x, y, 'o', color='black')

    return ax

def draw_line(ax, start, end, label):
    ax.plot([start[0], end[0]], [start[1], end[1]], label=label)

ax = draw_geoboard()

# Dibujamos algunas líneas como ejemplo y sus longitudes
draw_line(ax, (0,0), (1,0), "1")
draw_line(ax, (0,0), (0,1), "1")
draw_line(ax, (0,0), (1,1), r"$\sqrt{2}$")
draw_line(ax, (0,0), (2,0), "2")
#... Continúa con el resto de las líneas posibles ...

# Mostrar leyenda
ax.legend(loc="upper left")

plt.show()
```

```python
# Función para dibujar un geoboard
def draw_geoboard():
    for x in range(6):
        plt.plot([x]*6, range(6), 'bo-')
    for y in range(6):
        plt.plot(range(6), [y]*6, 'bo-')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlim(-1, 6)
    plt.ylim(-1, 6)

# Triángulo escaleno agudo
def acute_scalene():
    draw_geoboard()
    plt.plot([0, 4, 2, 0], [0, 1, 3, 0], 'ro-')
    plt.title("Acute Scalene Triangle")
    plt.show()

# Triángulo escaleno recto
def right_scalene():
    draw_geoboard()
    plt.plot([0, 0, 4, 0], [0, 3, 0, 0], 'ro-')
    plt.title("Right Scalene Triangle")
    plt.show()

# Triángulo escaleno obtuso
def obtuse_scalene():
    draw_geoboard()
    plt.plot([0, 4, 4, 0], [0, 0, 2, 0], 'ro-')
    plt.title("Obtuse Scalene Triangle")
    plt.show()

# Dibujar los triángulos
acute_scalene()
right_scalene()
obtuse_scalene()
```

```python
# Función para dibujar un geoboard
def draw_geoboard():
    for x in range(6):
        plt.plot([x]*6, range(6), 'bo-')
    for y in range(6):
        plt.plot(range(6), [y]*6, 'bo-')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlim(-1, 6)
    plt.ylim(-1, 6)

# Triángulo isósceles agudo
def acute_isosceles():
    draw_geoboard()
    plt.plot([1, 3, 2, 1], [0, 0, 4, 0], 'ro-')
    plt.title("Acute Isosceles Triangle")
    plt.show()

# Triángulo isósceles recto
def right_isosceles():
    draw_geoboard()
    plt.plot([0, 4, 0, 0], [0, 0, 4, 0], 'ro-')
    plt.title("Right Isosceles Triangle")
    plt.show()

# Triángulo isósceles obtuso
def obtuse_isosceles():
    draw_geoboard()
    plt.plot([0, 4, 2, 0], [1, 1, 4, 1], 'ro-')
    plt.title("Obtuse Isosceles Triangle")
    plt.show()

# Dibujar los triángulos
acute_isosceles()
right_isosceles()
obtuse_isosceles()
```
